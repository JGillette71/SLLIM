## Survey of data values

```python
import matplotlib.pyplot as plt

tactic_plot = log_data['MITRE ATT&CK Tactic'].value_counts().plot(kind="barh",
                                                                  title="MITRE ATT&CK Tactic Values",
                                                                  grid=True)

# Add count values to each bar
for index, value in enumerate(log_data['MITRE ATT&CK Tactic'].value_counts()):
    plt.text(value, index, str(value))
```

```python
technique_plot = log_data['MITRE ATT&CK Technique'].value_counts().plot(kind="barh",
                                                                  title="MITRE ATT&CK Technique Values",
                                                                  figsize = (8, 16),
                                                                  grid=True)

# Add count values to each bar
for index, value in enumerate(log_data['MITRE ATT&CK Technique'].value_counts()):
    plt.text(value, index, str(value))
```

```python
events_plot = log_data['Windows Event IDs'].value_counts().plot(kind="barh",
                                                                  title="Windows Event IDs Values",
                                                                  figsize = (8, 16),
                                                                  grid=True)

# Add count values to each bar
for index, value in enumerate(log_data['Windows Event IDs'].value_counts()):
    plt.text(value, index, str(value))
```

Depending on the Q&A pair generation strategy, semi-structured values in Windows Event IDs column may need to be normalized for atomicity, e.g. "7045 or 4697" sample transformed to two separate atomic samples.
